```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to load and plot feature importances from a CSV file
def plot_feature_importance(file_path, model_name):
    # Load feature importances from CSV
    feature_importance = pd.read_csv(file_path, index_col=0, header=None).squeeze("columns")
    
    # Plot feature importances
    plt.figure(figsize=(10, 8))
    sns.barplot(x=feature_importance.values, y=feature_importance.index)
    plt.title(f'{model_name} Feature Importance')
    plt.xlabel('Importance')
    plt.ylabel('Feature')
    plt.show()

# File paths and model names
file_paths = {
    "Random Forest": "feature_importance/random_forest_model.csv",
    "SVM": "feature_importance/svm_model.csv",
    "XGBoost": "feature_importance/xgboost_model.csv"
}

# Plotting for each model
for model_name, file_path in file_paths.items():
    plot_feature_importance(file_path, model_name)
```
