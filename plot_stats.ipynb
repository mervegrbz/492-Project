```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('data/combined_all.csv')

# Plot each feature with respect to the is_attack column
for column in data.columns:
    if column != 'is_attack':
        plt.figure(figsize=(10, 6))
        plt.title(f'Distribution of {column} by is_attack')
        data.boxplot(column=column, by='is_attack')
        plt.suptitle('')
        plt.xlabel('is_attack')
        plt.ylabel(column)
        plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('data/combined_all.csv')

# Plot each feature with respect to the is_attack column
for column in data.columns:
    if column != 'is_attack':
        plt.figure(figsize=(10, 6))
        sns.kdeplot(data=data, x=column, hue='is_attack', fill=True, common_norm=False, palette='crest', alpha=0.5)
        plt.title(f'Distribution of {column} by is_attack')
        plt.xlabel(column)
        plt.ylabel('Density')
        plt.legend(title='is_attack', labels=['Normal', 'Attack'])
        plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('data/combined_all.csv')

# Split the data into attack and normal flows
df_attack = data[data['is_attack'] == 1]
df_health = data[data['is_attack'] == 0]

# Convert the timestamp (Unnamed: 0) to a proper column name for clarity
data.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
df_attack.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
df_health.rename(columns={'Unnamed: 0': 'id'}, inplace=True)

# Plot each feature with respect to the is_attack column
for column in data.columns:
    if column not in ['is_attack', 'id']:
        plt.figure(figsize=(10, 6))
        plt.plot(df_attack['id'], df_attack[column], label='Attack')
        plt.plot(df_health['id'], df_health[column], label='Health')
        plt.xlabel('Timestamp')
        plt.ylabel(column)
        plt.title(f'{column} over time')
        plt.legend()
        plt.show()
```
